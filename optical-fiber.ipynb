```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib widget
# from ipywidgets import interact
from scipy.special import jv, yn
```

```python
x = np.linspace(3, 20, N:=100)
v = 10
y = np.sqrt(v**2 - x**2)
l = 5


xp = np.linspace((x[1]+x[0])/2, (x[-1]+x[-2])/2, N-1)

J = jv(l, x)
Jp = np.diff(J) / (x[1]-x[0])
Jp_rec = - jv(l+1, x) + l * J / x


K = yn(l, x)
Kp = np.diff(K) / (x[1]-x[0])
Kp_rec = + yn(l-1, x) - l * K / x

#plt.plot(x, J)
#plt.plot(x, Jp_rec)
#plt.plot(xp, Jp)

plt.plot(x, K)
plt.plot(x, Kp_rec)
plt.plot(xp, Kp)

plt.xlim(0, v)
# plt.ylim(-3, 3)
```

```python
x = np.linspace(0, 15, 2000)
v = 3
y = np.sqrt(v**2  - x**2)

R = x * jv(1, x) / jv(0, x)
L = y * yn(1, y) / yn(0, y)

plt.plot(x, R, ls="", marker=".", ms=1)
plt.plot(x, L, ls="", marker=".", ms=1)

plt.ylim(0, 50)
```
